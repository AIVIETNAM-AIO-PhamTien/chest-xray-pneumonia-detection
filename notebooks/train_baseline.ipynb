## 5. Chạy training

Ghi đè `data.root_dir` trong `configs/baseline.yaml` bằng `DATA_ROOT` đã xác định ở bước 2,
rồi chạy training.

```python
import sys

import yaml

sys.path.append(".")

with open("configs/baseline.yaml") as f:
    cfg_dict = yaml.safe_load(f)
cfg_dict["data"]["root_dir"] = DATA_ROOT
with open("configs/baseline.yaml", "w") as f:
    yaml.safe_dump(cfg_dict, f)

from src.train import run_training

run_training("configs/baseline.yaml")
```
